```julia
function aproximacirculo(r)
    n1=0
    n2=0
    squares=0
    contador=Float64(0)
    for n1 in 0:1:r
        for n2 in 0:1:r
            squares=n1*n1+n2*n2
            if squares<=r*r
                contador+=1
            end
        end
    end
    return contador
end
```

```julia
aproximacirculo(0)
```

```julia
function aproximaciones(rmax,n)
    X=zeros(0)
    R=zeros(0)
    for r in 0:rmax/n:rmax
        v=aproximacirculo(r)
        push!(X,v)
        push!(R,r)
    end
    return R,X
end
        
```

```julia
R,X=aproximaciones(10000,100)
```

```julia
using PyPlot
```

```julia
plot(R,.25.*pi.*R.*R,color="r", label=L"$A(R)=\frac{1}{4}\pi R^{2}$")
plot(R,X,color="green", linestyle="dashed", marker="o",
markerfacecolor="green", markersize=2.5, label=L"$A(R):\, n_{1}^{2}+n_2^{2}\leq R^{2}$")
legend(loc=2)
xlabel(L"$R$")
ylabel(L"$A(R)$")
title(L"Aproximación al área para $n_{1},n_{2}\geq 0$")
```

```julia
function aproximacirculo2(r)
    n1=0
    n2=0
    squares=0
    contador=Float64(0)
    for n1 in -r:1:r
        for n2 in -r:1:r
            squares=n1*n1+n2*n2
            if squares<=r*r
                contador+=1
            end
        end
    end
    return contador/4
end
```

```julia
function aproximaciones2(rmax,n)
    X=zeros(0)
    R=zeros(0)
    for r in 0:rmax/n:rmax
        v=aproximacirculo2(r)
        push!(X,v)
        push!(R,r)
    end
    return R,X
end
        
```

```julia
R,X=aproximaciones2(10,10)
```

```julia
plot(R,.25.*pi.*R.*R,color="r", label=L"$A(R)=\frac{1}{4}\pi R^{2}$")
plot(R,X,color="green", linestyle="dashed", marker="o",
markerfacecolor="green", markersize=2.5, label=L"$A(R):\, n_{1}^{2}+n_2^{2}\leq R^{2}$")
legend(loc=2)
xlabel(L"$R$")
ylabel(L"$A(R)$")
title(L"Aproximación al área para $n_{1},n_{2}\in \mathbb{Z}$")
```
