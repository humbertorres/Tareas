# Cálculo de $h$ y $\phi$

```julia
using LsqFit
```

```julia
model(x,p)=p[1]*x.+p[2]
```

```julia
ν=[ν[1],ν[2],ν[3],ν[4],ν[5],ν[6]]
vf=[vf[1],vf[2],vf[3],vf[4],vf[5],vf[6]]
```

```julia
#Obtenemos los valores aproximados del modelo lineal
htest,Φtest=linreg(ν,vf)
```

```julia
#Hacemos un ajuste de mínimos cuadrados, partiendo de los valores aproximados obtenidos
fit=curve_fit(model,ν,vf,[htest,Φtest])
```

```julia
#Calculamos la pendiente y ordenada del ajuste realizado
h,Φ=fit.param
```

```julia
#y sus errores, con un intervalo de confianza del 90%
δh,δΦ=estimate_errors(fit,.90)
```

```julia
errorbar(ν,vf,δvf2, fmt="o", c="r", markersize=5, label=L"V_{f}\, experimentales")
plot(ν,model(ν,[h,Φ]), label="Ajuste lineal", c="k")
annotate(L"h=1.851\pm .266\times 10^{-15}eV,\, \Phi=.496\pm.177 eV",(6.3E14,-.65))
legend()
xlabel(L"ν(Hz)")
ylabel(L"V_{f} (Volts)")
```
